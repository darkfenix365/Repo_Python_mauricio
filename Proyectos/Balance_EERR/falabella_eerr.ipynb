```python
# Importacion de librerias a ocupar
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from parametros import TIKR_MAIL, TIKR_PASSWORD
import pandas as pd
import numpy as np
from time import sleep
import tempfile
```

```python
temp = tempfile.TemporaryDirectory("w+t")
tempdir = temp.name

# Generacion de driver chrome
def create_driver():    
    # Chrome options
    options = webdriver.ChromeOptions()
    options.add_argument('--no-sandbox')
    download_argument = f'download.default_directory={tempdir}'
    print(download_argument)
    prefs = {'download.default_directory': f'{tempdir}'}
    options.add_experimental_option('prefs', prefs)
    # Conexccion a chrome driver
    driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
    driver.maximize_window()
    driver.get('https://app.tikr.com/login')
    return driver

driver = create_driver()

#Ingreso a la plataforma
sleep(5)
driver.find_element("xpath", '/html/body/div/div/div/div/main/div/div/div/div/div/div/div/div[3]/div[1]/div/div[1]/div/input').send_keys(TIKR_MAIL)
driver.find_element("xpath", '/html/body/div/div/div/div/main/div/div/div/div/div/div/div/div[3]/div[2]/div/div/div[1]/input').send_keys(TIKR_PASSWORD)
driver.find_element("xpath", '/html/body/div/div/div/div/main/div/div/div/div/div/div/div/div[3]/button').click()
sleep(10)

# Ingreso de falabella 
driver.get('https://app.tikr.com/stock/financials?cid=882992&tid=20189944&tab=is&ref=mbqiez')
sleep(30)
driver.find_element("xpath", '/html/body/div/div/div/div[1]/main/div/div/div/div/div/div/div[4]/div/div/div/div[1]/div/div/div/div/div[2]/div[1]/div[4]/div/button').click()
sleep(10)
driver.close()
```
